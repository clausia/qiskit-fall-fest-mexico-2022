# Entrelazamiento cuántico

Dos partículas están **entrelazadas** cuando juntas forman un *sistema cuántico*, independientemente de la distancia entre ellas.

Para entender este tipo de conexión cuántica, consideremos dos electrones. Los electrones tienen una propiedad cuántica llamada [espín](https://en.wikipedia.org/wiki/Spin_(physics)) que, cuando se mide, puede tomar uno de dos valores, denominados "arriba" o "abajo". Medir el espín de cada electrón es como lanzar una moneda al aire: saldrá aleatoriamente hacia arriba o hacia abajo.

<p style="font-size:10px;font-style:italic;text-align:right">Imagen de: <a href="https://www.chemistrylearner.com/spin-quantum-number.html" target="_blank">Chemistry Learner</a></p>

Ahora imaginemos que dos amigos, Alice y Bob, reciben cada uno una serie de monedas por correo. A medida que llega cada par de monedas, los amigos las lanzan al mismo tiempo. Alice podría obtener la secuencia cara, cruz, cruz, cara, cruz. Y Bob podría obtener cara, cara, cruz, cruz, cruz. Los resultados de los lanzamientos de moneda de Alice y Bob no tendrán nada que ver entre sí.

Pero si repiten este experimento con una serie de **electrones enredados** en lugar de monedas, obtendrán un resultado extraño: cada vez que Alice mida un electrón con **espín-arriba**, Bob encontrará que su mitad correspondiente del par de electrones tiene **espín-abajo**, y viceversa. Los dos actos de medición están conectados, casi como si lanzar una moneda al aire pudiera enviar una señal que asegurara instantáneamente el resultado correspondiente de su compañero distante en el momento preciso de la medición.

Fue [Albert Einstein](https://en.wikipedia.org/wiki/Albert_Einstein), junto con [Boris Podolsky](https://en.wikipedia.org/wiki/Boris_Podolsky) y [Nathan Rosen](https://en.wikipedia.org/wiki/Nathan_Rosen), quienes describieron por primera vez sobre el entrelazamiento cuántico en un [artículo de 1935](https://cds.cern.ch/record/405662/files/PhysRev.47.777.pdf). El fenómeno, cuyos efectos Einstein denominó despectivamente "acción espeluznante a distancia", fue una consecuencia inevitable de la naciente teoría de la mecánica cuántica. Einstein sospechó que el entrelazamiento sería la *sentencia de muerte* de la mecánica cuántica porque *parecía* ir en contra de un principio central de la relatividad: que ninguna información podría viajar más rápido que la velocidad de la luz. Ninguna medida de un electrón debería poder influir instantáneamente en una medida en algún lugar distante.

En cambio, su artículo sentaría las bases para un replanteamiento completo de la realidad y un nuevo y radical campo de investigación. [[QuantaMagazine, 2022](https://www.quantamagazine.org/pioneering-quantum-physicists-win-nobel-prize-in-physics-20221004/)]

# Un poco de historia

- 1935\
  Einstein, Podolsky y Rosen, argumentan que a la física cuántica le hacen falta algunas "variables ocultas" que explicarían aparentes paradojas cuánticas.


- 1964\
  **John Bell** diseña un experimento mental que sugiere que los estados de las partículas entrelazadas están demasiado correlacionados para ser explicados por variables ocultas.


- 1969\
  El físico estadounidense **John Clauser** y sus colegas refinan el teorema de Bell y proponen un experimento para probarlo con datos del mundo real.


- 1972\
  **Clauser** y un colega observan un inclumplimiento de la desigualdad de Bell por primera vez en un experimento.


- 1982\
  **Alain Aspect** de Francia lleva a cabo una serie de experimentos que confirman los incumplimientos del teorema de Bell, pero quedan algunas lagunas.


- 2017\
  El físico australiano **Anton Zeilinger** y sus colegas demuestran una prueba sin lagunas que finalmente cierra la puerta a la teoría de variables ocultas de Einstein, Podolsky y Rosen.


- 2022\
  **Clauser**, **Aspect** y **Zeilinger** reciben el **Premio Nobel de Física** por sus investigaciones sobre el entrelazamiento cuántico ([press release](https://www.nobelprize.org/prizes/physics/2022/press-release/))

# Teorema de Bell

<a id="3.1"></a>
## El problema

El Teorema de Bell es uno de los resultados más importantes en la ciencia moderna. Permite diseñar un experimento que comprueba que los **procesos fundamentales de la naturaleza son aleatorios**.

¿Qué tipo de procesos? Por ejemplo un rayo de luz que pasa por un *[divisor de haz](https://en.wikipedia.org/wiki/Beam_splitter)* (o *beam splitter* en inglés), la mitad de la luz es reflejada y la otra mitad es transmitida. Pero si consideramos a la luz fotón a fotón, ¿cómo saber cuál será reflejado y cuál será transmitido?. La física cuántica, dice que es **puramente aleatorio**.

<p style="font-size:10px;font-style:italic;text-align:right">Imágenes de: <a href="https://www.edmundoptics.com/knowledge-center/application-notes/optics/what-are-beamsplitters/" target="_blank">(izquierda) Edmund Optics</a> y <a href="https://www.quantamagazine.org/a-new-tool-for-finding-dark-matter-digs-up-nothing-20220321/" target="_blank">(derecha) Quanta Magazine</a></p>

Albert Einstein no creia en esta aleatoriedad, él decía que debían haber **variables ocultas**: una especie de código secreto escrito dentro de las particulas, que diactaba el comportamiento de cada fotón.

De acuerdo a la teoría de *variables ocultas*: hay un número (una vaiable oculta), que contiene toda la información (previo al experimento), sobre qué fotón en particular hará cuando llegue al *beam spliter*.

$$$$

*Pero, ¿esta teoría es correcta? ¿Este número realmente existe?*

**¿Cómo saberlo? si está oculto.**

$$$$

Precisamente el **Teorema de Bell** provee un método para saber si las *variables ocultas* existen.

<a id="3.2"></a>
## Razonamiento

Ahora veamos las bases del Teorema de Bell, utilizando la lógica.

*Recordemos*, la pregunta es **¿la variable oculta realmente existe?**

El problema es que no es posible probar que un número particular no existe, quizá solo está oculto.

Pero si definimos ciertas condiciones lógicas, entonces podemos probar que $4$ números particulares no pueden existir al mismo tiempo.

$$$$

Supongamos que $A$, $B$, $C$, $D$ son valores reales que existen simultáneamente. Luego, ya que los valores están determinados, entonces podemos compararlos y ver si son iguales o no:


|    ✔️   |   ✔️   |    ✔️   |        |
|:-------:|:-------:|:-------:|:-------:|
| $$~~~~~~~~~A=B~~~~~~~~~$$ | $$~~~~~~~~~B=C~~~~~~~~~$$ | $$~~~~~~~~~C=D~~~~~~~~~$$ | $$~~~~~~~~~A=D~~~~❓~~~~$$  |
|         |        |         |    ✔️  |


Ahora, si negamos la cuarta igualdad

|         |         |         |         |
|:-------:|:-------:|:-------:|:-------:|
| $$~~~~~~~~~A=B~~~~~~~~~$$ | $$~~~~~~~~~B=C~~~~~~~~~$$ | $$~~~~~~~~~C=D~~~~~~~~~$$ | $$~~~~~~~~~A\color{red}{\neq} D~~~~~~~~$$  |

Entonces, claramente, estas 4 igualdades no pueden ser verdad al mismo tiempo para cuatro valores particulares.

Así, hemos definido una **condición** para la existencia de estos cuatro números.

$$$$

Reescribamos la condición en forma matemática:

Como solo importa si los valores $A$, $B$, $C$, $D$ son iguales o diferentes, solo basta con considerar 2 posibles valores, escogemos $+1$ y $-1$, por ejemplo:

$$
\begin{aligned}
A&=+1\\
B&=-1\\
C&=+1\\
D&=-1
\end{aligned}$$

Comparando las variables con estos valores de ejemplo, podemos notar que las comparaciones, como se definieron, serán verdaderas o falsas:

|         |         |         |         |
|:-------:|:-------:|:-------:|:-------:|
| $$~~~~~~~~~A=B~~~~~~~~~$$ | $$~~~~~~~~~B=C~~~~~~~~~$$ | $$~~~~~~~~~C=D~~~~~~~~~$$ | $$~~~~~~~~~A\color{red}{\neq}D~~~~~~~~$$  |
|    ❌   |   ❌   |    ❌   |    ✔️   |

$$$$

De este modo, para todos los posibles valores (tomados del conjunto $\{+1, -1\}$) que las variables pueden tomar, tendremos que la cantidad de comparaciones verdaderas menos la cantidad de comparaciones falsas estará acotada con:

$$$$
$$-2\leq \color{green}{V}-\color{red}{F}\leq 2$$
$$$$

En el ejemplo anterior tenemos que la cantidad de comparaciones verdaderas es $\color{green}{V}=1$, mientras que la cantidad de comparaciones falsas es $\color{red}{F}=3$, entonces $\color{green}{V}-\color{red}{F}=1-3=-2$, lo que cumple con la condición de acotamiento.
$$$$

Notemos que $\color{green}{V}$ y $\color{red}{F}$ no son objetos matemáticos.

Tomemos el ejemplo cuando:

$$
\begin{aligned}
A&=+1\\
B&=+1\\
C&=+1\\
D&=-1
\end{aligned}$$

Entonces, asignaremos un $\color{green}{+1}$ a las comparaciones que son verdaderas $\color{green}{V}$, y asignaremos un $\color{red}{-1}$ a las que son falsas $\color{red}{F}$, llamándolo **valor de verdad**, entonces las comparaciones se verían como:

|         |         |         |         |
|:-------:|:-------:|:-------:|:-------:|
| $$~~~~~~~~~A=B~~~~~~~~~$$ | $$~~~~~~~~~B=C~~~~~~~~~$$ | $$~~~~~~~~~C=D~~~~~~~~~$$ | $$~~~~~~~~~A\color{red}{\neq}D~~~~~~~~$$  |
|    $\color{green}{+1}$   |   $\color{green}{+1}$   |    $\color{red}{-1}$   |    $\color{green}{+1}$   |

Así, el acotamiento se vería como:

$$$$
$$-2\leq \text{suma de los valores de verdad} \leq 2$$
$$$$

Notemos que la ventaja de haber escogido que los valores posible de las variables $A$, $B$, $C$, $D$ sea uno de los posibles elementos en $\{+1, -1\}$, es que podemos efectuar el **producto** de ambas variables involucradas en una comparación, para obtener el *valor de verdad* correcto, es decir:

|         |         |         |         |
|:-------:|:-------:|:-------:|:-------:|
| $$~~~~~~~~~A=B~~~~~~~~~$$ | $$~~~~~~~~~B=C~~~~~~~~~$$ | $$~~~~~~~~~C=D~~~~~~~~~$$ | $$~~~~~~~~~A\color{red}{\neq}D~~~~~~~~$$  |
|    $\color{green}{+1}$   |   $\color{green}{+1}$   |    $\color{red}{-1}$   |    $\color{green}{+1}$   |
| $$~~~~~~~~~A\cdot B~~~~~~~~~$$ | $$~~~~~~~~~B\cdot C~~~~~~~~~$$ | $$~~~~~~~~~C\cdot D~~~~~~~~~$$ | $$~~~~~~~~~\color{red}{-}A\cdot D~~~~~~~~$$  |

El signo menos enfrente de $\color{red}{-}A\cdot D$ es porque para esa comparación estamos validando que $A$ no sea igual a $D$.

$$$$

Ahora si, tenemos una condición matemática que puede ser calculada:

$$$$
$$\boxed{-2 \; \; \leq \; \; A\cdot B+B\cdot C+C\cdot D \color{red}{-}A\cdot D \; \; \leq \; \; 2}$$
$$$$


> Si las variables $A$, $B$, $C$, $D$ toman valores reales que existen simultáneamente (del conjunto $\{+1, -1\}$), entonces la desigualdad anterior debe mantenerse (para cualquier combinación de dichos valores).

<a id="3.3"></a>
## Desigualdad de Bell

Recordemos que la pregunta que buscamos responder es:

> ¿Los procesos fundamentales de la naturaleza son verdaderamente **aleaorios**?

$$$$

Consideremos un escenario simple para un proceso aleatrio: Una partícula (una moneda en nuestra analogía) en una máquina tragamonedas con dos posibles resultados en el conjunto $\{🍒, 🍋\}$.

¿El comportamiento de la partícula es realmente aleatorio?
- En el sentido de que el resultado no esté predeterminado

¿O existen una variable oculta $A$ con un valor dado desde el inicio?
- De tal manera que si $A=1$ entonces obtenemos un resultado (🍒) y si $A=-1$ obtenemos el otro (🍋)?

Pero no podemos probar que algo no existe. Quizá la variable $A$ esté oculta junto con el pegaso.

De hecho, dados los resultados de la máquina tragamonedas, podemos generar una lista de valores para la variable oculta $A$ que reproduce el comportamiento de la partícula, por lo que podemos creer que dicha lista existía antes del experimento.

Entonces, no podemos descartar la existencia de *una sola variable oculta*.

$$$$

Aquí vienen la genialidad de **[John Stewart Bell](https://en.wikipedia.org/wiki/John_Stewart_Bell)**: diseñó un experimento que podía descartar la existencia de $4$ variables ocultas $\rightarrow$ **[Prueba de Bell](https://en.wikipedia.org/wiki/Bell_test)**.

Por cada partícula, debemos elegir entre dos máquinas tragamonedas, la de arriba o la de abajo, y esa elección es aleatoria, decisión tomada, por ejemplo, al tirar por un dado.

$$$$

En la prueba de Bell comenzamos suponiendo que **si** existen variables ocultas y el resultado de la prueba demostrará si esta suposición es correcta o incorrecta. 

Supongamos que las variables ocultas $A$, $B$, $C$, $D$, tienen valores definidos antes de que las partículas sean enviadas a las máquinas tragamonedas, si este es el caso, entonces las variables ocultas satisfacen la desigualdad que encontramos antes,

$$$$
$$-2 \; \; \leq \; \; A\cdot B+B\cdot C+C\cdot D -A\cdot D \; \; \leq \; \; 2$$
$$$$

Y deben satisfacer esta desigualdad para cualquier valor que puedan tomar, por lo tanto la desigualdad se debe mantener también para el **valor promedio** de las variables ocultas.

$$$$
$$-2 \; \; \leq \; \; \langle \, A\cdot B+B\cdot C+C\cdot D -A\cdot D \, \rangle \; \; \leq \; \; 2$$
$$$$

Notemos que para **cada ronda** solo obtendremos un posible producto, ya sea,

$$A\cdot B$$ ó $$B\cdot C$$ ó $$C\cdot D$$ ó $$D\cdot A$$

Entonces, necesitamos muchas rondas para probar la desigualdad, por lo que llevamos a cabo el experimentos múltiples veces para minimizar posibles errores de medición.

$$$$

Entonces la desigualdad que podemos probar con este experimento es usando promedios en lugar de medidas aisladas, por lo que quedaría de esta manera:

$$$$
$$-2 \; \; \leq \; \; \langle A\cdot B\rangle+\langle B\cdot C\rangle+\langle C\cdot D\rangle-\langle A\cdot D\rangle \; \; \leq \; \; 2$$
$$$$

> A esto se le llama la **Desigualdad de Bell (Bell Inequality)**.

La prueba de Bell se ha realizado muchas veces, desde los 80's y en muchos laboratorios alrededor del mundo, y gracias a las propiedades cuánticas de las partículas y a la ingeniosa preparación de los escenarios de experimentación, muchas veces se ha encontrado que el resultado de los valores promedio es **¡mayor que 2!**

$$$$
$$\boxed{\lvert \langle A\cdot B\rangle+\langle B\cdot C\rangle+\langle C\cdot D\rangle-\langle A\cdot D\rangle \rvert \; \; \color{darkorange}{>} \; \; 2}$$
$$$$

> Esto es conocido como el **incumplimiento de la desigualdad de Bell**.

$$$$

Este incumplimiento demuestra que $A$, $B$, $C$, $D$ no fueron determinadas antes del experimento.

De hecho, al suponer que existen variables ocultas implica que es válida la desigualdad de Bell, entonces el que no se cumpla la desigualdad implica que **no existen variables ocultas**.

Por lo que, las variables ocultas son descartadas por un experimento de Bell exitoso.

$$\lvert \langle A\cdot B\rangle+\langle B\cdot C\rangle+\langle C\cdot D\rangle-\langle A\cdot D\rangle \rvert \; \; \color{darkorange}{>} \; \; 2$$

$$$$

<a id="3.4"></a>
## Lagunas

Una prueba de Bell exitosa es una implicación profunda para la física, la filosofía y la tecnología, por lo que debemos asegurarnos que la implementación experimental realmente respete las suposiciones del teorema de Bell, si no lo hace, se producirá una laguna.

Hay muchas posibles defectos experimentales que pueden afectar la rigurosidad de la demostración, estas lagunas apuntan a en los límites de los métodos experimentales y desafían directamente las habilidades técnicas de los grupos de investigación.

$$$$

Una de esas lagunas fue abordada en el experimento de 2016 llamado "[The Big Bell Test](http://thebigbelltest.icfo.eu/)".

Recordemos el dado en nuestro experimento, la decisión de elegir la máquina tragamonedas (arriba o abajo) está decidida por un dado, el reto es que si cada particula tiene acceso a la elección de su dado, nuestra demostración se rompe y no podemos demostrar la existencia de variables ocultas.

Considera una de las partículas, la prueba de Bell supone que tiene dos variables ocultas $A$ y $C$, si una partícula encuentra una manera de conocer las opciones de los dados, entonces se pueden usar para construir una lista única de valores ocultos.

El problema se reduce a suponer una sola variable oculta para cada partícula, entonces nuestra prueba que necesita 4 variables ocultas ¡se rompe!

Entonces un elemento crucial en la prueba de Bell es garantizar que las elecciones de las máquinas tragamonedas (el resultado de los dados) no sean accesibles por las partículas, esta es la laguna de la **libertad de elección**.

Entonces ¿Cuál es el mejor sistema de elección? No existe respuesta para eso. Cada equipo de investigación tiene su propio sistema favorito de elección.

El sistema favorito de elección del equipo de "The Big Bell Test" fue: la gente, llamada la mejor prueba, las personas eligieron los valores de los dados, ayudando a probar las desigualdades de Bell en 12 laboratorios diferentes alrededor del mundo en el mismo día. Los resultados de la prueba, realizada el 30 de noviembre de 2016, se publicaron en la revista Nature, si quieres saber más de esta prueba puedes consultarla en http://thebigbelltest.org/.

# Desigualdad CHSH

La **desigualdad CHSH** puede usarse en la demostración del teorema de Bell, que establece que ciertas consecuencias del entrelazamiento en la mecánica cuántica no pueden reproducirse mediante teorías locales de variables ocultas. 

La verificación experimental de la desigualdad que se incumple, se ve como una confirmación de que la naturaleza no puede ser descrita por tales teorías. CHSH significa **John Clauser**, **Michael Horne**, **Abner Shimony** y **Richard Holt**, quienes lo describieron en un [artículo](https://doi.org/10.1103%2FPhysRevLett.23.880) muy citado publicado en 1969.

Derivaron la desigualdad CHSH, que, al igual que con la desigualdad original de Bell, es una restricción sobre la ocurrencia estadística de "coincidencias" en una prueba de Bell que es necesariamente cierta si existen variables ocultas locales subyacentes, una suposición que es a veces llamado [realismo local](https://en.wikipedia.org/wiki/Local_realism). De hecho, la desigualdad se deja de cumplir rutinariamente en los experimentos modernos de mecánica cuántica [Markoff, J, 2015](https://www.nytimes.com/2015/10/22/science/quantum-theory-experiment-said-to-prove-spooky-interactions.html).

La forma usual de la [desigualdad CHSH](https://en.wikipedia.org/wiki/CHSH_inequality) es,

$$|S|\leq 2$$

donde,

$$S=E(a,b)-E(a,b')+E(a',b)+E(a',b')$$

$a$ y $a'$ son configuraciones del detector en el lado $A$, $b$ y $b'$ en el lado $B$, las cuatro combinaciones se prueban en subexperimentos separados. Los términos $E(a, b)$, etc. son las [correlaciones cuánticas](https://en.wikipedia.org/wiki/Quantum_correlation) de los pares de partículas, donde la correlación cuántica se define como el valor esperado del producto de los "resultados" del experimento, es decir, el promedio estadístico de $A(a)\cdot B(b)$, donde $A$ y $B$ son los resultados separados, usando la codificación $+1$ para el canal '+' y $-1$ para el canal '−'.

La derivación de Clauser, et al. de 1969, estaba orientada hacia el uso de detectores de *dos canales* y, de hecho, es para estos que generalmente se usa, pero según su método, los únicos resultados posibles eran $+1$ y $-1$. Para adaptarse a situaciones reales, que en ese momento significaban el uso de luz polarizada y polarizadores de un solo canal, tuvieron que interpretar '-' como "no detección en el canal '+'", es decir, '-' o nada. En el artículo original no discutieron cómo se podría aplicar la desigualdad de dos canales en experimentos reales con detectores imperfectos reales, aunque más tarde se demostró [ref](https://en.wikipedia.org/wiki/CHSH_inequality#cite_note-Bell-1971-4) que la desigualdad en sí misma era igualmente válida. Sin embargo, la ocurrencia de resultados cero significa que ya no es tan obvio cómo se deben estimar los valores de $E$ a partir de los datos experimentales.

El formalismo matemático de la mecánica cuántica predice un valor máximo para $S$ que es de $2\sqrt{2}\approx 2.8$ ([límite de Tsirelson](https://en.wikipedia.org/wiki/Tsirelson%27s_bound)), que es mayor que $2$ y, por lo tanto, la teoría de la mecánica cuántica predice el incumplimiento de la desigualdad de CHSH.

En la práctica, la mayoría de los experimentos reales han utilizado luz en lugar de los electrones que Bell originalmente tenía en mente. La propiedad de interés es, en los experimentos más conocidos, la **dirección de polarización**, aunque se pueden utilizar otras propiedades. El diagrama muestra un experimento óptico típico. Se registran las coincidencias (detecciones simultáneas), los resultados se clasifican como '$++$', '$+-$', '$-+$' o '$--$' y se acumulan los recuentos correspondientes.

$$$$

Se llevan a cabo cuatro subexperimentos separados, correspondientes a los cuatro términos $E(a,b)$ en la estadística de prueba $S$. En la práctica, generalmente se eligen los ajustes $a = 0^{\circ}$, $a' = 45^{\circ}$, $b = 22.5^{\circ}$ y $b' = 67.5^{\circ}$, llamados los **ángulos de prueba de Bell**, siendo estos aquellos para los que la fórmula mecánica cuántica da el mayor incumplimiento de la desigualdad.

Para cada valor seleccionado de $a$ y $b$, el número de coincidencias en cada categoría $\{N_{++}, N_{--},N_{+-},N_{-+}\}$ son registrados. La estimación experimental para $E(a,b)$$ se calcula como:

$$E=\frac{N_{++}-N_{+-}-N_{-+}+N_{--}}{N_{++}+N_{+-}+N_{-+}+N_{--}}$$

Podemos usar un laboratorio virtual para comprobar el incumplimiento de la desigualdad CHSH. La configuración, realizada en [Quantum Flytrap](https://quantumflytrap.com/), puede ser encontrada [aquí](https://lab.quantumflytrap.com/lab/bell-inequality?mode=beam).

# Desigualdad CHSH con Qiskit

Importamos las librerías necesarias.

```python
import qiskit
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile, Aer, IBMQ
from qiskit.tools.visualization import circuit_drawer
from qiskit.tools.monitor import job_monitor, backend_monitor, backend_overview
from qiskit.providers.aer import noise

import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# usaremos una computadora cuántica real en la nube

IBMQ.load_account()

provider = IBMQ.get_provider('ibm-q')
quito = provider.get_backend('ibmq_quito')
```

```python
# también usaremos un simulador local

sim = Aer.get_backend('aer_simulator')
```

Primero vamos a definir una función para crear nuestros circuitos CHSH. Vamos a elegir, sin pérdida de generalidad, que Bob siempre use las bases: computacional ($Z$) y $X$ para sus medidas $B$ y $b$, respectivamente; mientras que Alice elige también bases ortogonales pero cuyo ángulo vamos a variar entre $0$ y $2\pi$ con respecto a las bases de Bob. Este conjunto de ángulos será el argumento de entrada para nuestra función de construcción de circuitos CHSH.

```python
def make_chsh_circuit(theta_vec):
    """Devuelve una lista de circuitos cuánticos para usar en un experimento CHSH
    (uno para cada valor de theta en theta_vec)
    
        Args:
            theta_vec (list): lista de valores de ángulos entre las bases de Alice y Bob
        
        Returns:
            List[QuantumCircuit]: CHSH QuantumCircuits para cada valor de theta
    """
    chsh_circuits = []
    
    for theta in theta_vec:
        obs_vec = ['00', '01', '10', '11']
        for el in obs_vec:
            qc = QuantumCircuit(2,2)
            qc.h(0)
            qc.cx(0, 1)
            qc.ry(theta, 0)
            for a in range(2):
                if el[a] == '1':
                    qc.h(a) 
            qc.measure(range(2),range(2))
            chsh_circuits.append(qc)

    return chsh_circuits
```

A continuación, vamos a definir una función para estimar la cantidad $S=\langle CHSH\rangle$. Podemos definir dos de tales cantidades, en realidad, 

$$\langle CHSH_1\rangle=\langle AB\rangle-\langle Ab\rangle+\langle aB\rangle+\langle ab\rangle$$

y

$$\langle CHSH_2\rangle=\langle AB\rangle+\langle Ab\rangle-\langle aB\rangle+\langle ab\rangle$$

Una vez elegidos los ejes de medición correspondientes para ambas partes, cada valor esperado puede estimarse simplemente sumando los recuentos de las cadenas de bits de salida con el signo apropiado (sumando los términos pares $00$ y $11$ y restando los términos impares $01$ y $10$).

```python
def compute_chsh_witness(counts):
    """Calcula los valores esperados para la desigualdad CHSH,
    para cada ángulo (theta) entre los ejes de medición.

        Args: counts (list[dict]): diccionario de conteos para cada experimento
              (4 por valor de theta)

        Returns:
            Tuple(List, List): Tupla de listas con los dos testigos de CHSH
    """
    # El orden es: ZZ,ZX,XZ,XX
    
    CHSH1 = []
    CHSH2 = []
    
    # Dividir la lista de diccionarios en conjuntos de 4
    for i in range(0, len(counts), 4):  
        theta_dict = counts[i:i + 4]
        zz = theta_dict[0]
        zx = theta_dict[1]
        xz = theta_dict[2]
        xx = theta_dict[3]

        no_shots = sum(xx[y] for y in xx)

        chsh1 = 0
        chsh2 = 0

        for element in zz:
            parity = (-1)**(int(element[0])+int(element[1]))
            chsh1+= parity*zz[element]
            chsh2+= parity*zz[element]

        for element in zx:
            parity = (-1)**(int(element[0])+int(element[1]))
            chsh1+= parity*zx[element]
            chsh2-= parity*zx[element]

        for element in xz:
            parity = (-1)**(int(element[0])+int(element[1]))
            chsh1-= parity*xz[element]
            chsh2+= parity*xz[element]

        for element in xx:
            parity = (-1)**(int(element[0])+int(element[1]))
            chsh1+= parity*xx[element]
            chsh2+= parity*xx[element]

        CHSH1.append(chsh1/no_shots)
        CHSH2.append(chsh2/no_shots)
    
    return CHSH1, CHSH2
```

Finalmente, vamos a dividir el intervalo $[0,2\pi)$ en $15$ ángulos y construiremos el conjunto correspondiente de circuitos CHSH.

```python
number_of_thetas = 15
theta_vec = np.linspace(0,2*np.pi,number_of_thetas)
my_chsh_circuits = make_chsh_circuit(theta_vec)
```

```python
my_chsh_circuits[4].draw('mpl')
```

```python
my_chsh_circuits[10].draw('mpl')
```

Estos circuitos simplemente crean un par de Bell y luego miden cada parte en una base diferente. Mientras que Bob ($q_1$) siempre mide en la base computacional o en la base X, la base de medición de Alice gira en el ángulo $\theta$ con respecto a la de Bob.

```python
# Ejecutar y obtener conteos

result_ideal = sim.run(my_chsh_circuits).result()

tic = time.time()
transpiled_circuits = transpile(my_chsh_circuits, quito)
job_real = quito.run(transpiled_circuits, shots=8192)
job_monitor(job_real)
result_real = job_real.result()
toc = time.time()

print(toc-tic)
```

```python
CHSH1_ideal, CHSH2_ideal = compute_chsh_witness(result_ideal.get_counts())
CHSH1_real, CHSH2_real = compute_chsh_witness(result_real.get_counts())
```

```python
plt.figure(figsize=(12, 8))
plt.rcParams.update({'font.size': 22})
plt.plot(theta_vec, CHSH1_ideal,'o-', label = 'CHSH1 Sin ruido')
plt.plot(theta_vec, CHSH2_ideal,'o-', label = 'CHSH2 Sin ruido')

plt.plot(theta_vec, CHSH1_real,'x-', label = 'CHSH1 Quito')
plt.plot(theta_vec, CHSH2_real,'x-', label = 'CHSH2 Quito')

plt.grid(which='major',axis='both')
plt.rcParams.update({'font.size': 16})
plt.legend()
plt.axhline(y=2, color='r', linestyle='-')
plt.axhline(y=-2, color='r', linestyle='-')
plt.axhline(y=np.sqrt(2)*2, color='k', linestyle='-.')
plt.axhline(y=-np.sqrt(2)*2, color='k', linestyle='-.')
plt.xlabel('Theta')
plt.ylabel('Testigo CHSH')
```

Observemos lo que sucedió. Hay algunas combinaciones particulares de bases de medición para las cuales $|S|\geq 2$.

¿Por qué?
Miremos nuestro sistema entrelazado bipartito. Es fácil demostrar que si

$$\ket{\psi}=\frac{1}{\sqrt2}\big(\ket{00}+\ket{11}\big)$$

entonces el valor esparado

$$\langle AB\rangle = \langle \psi |A\otimes B|\psi\rangle=-\cos\theta_{AB}$$

es el ángulo entre las bases de medición $A$ y $B$. Por lo tanto, para la elección particular de las bases $A=\frac{1}{\sqrt2}\big(\sigma_z-\sigma_x\big)$ y $a=\frac{1}{\sqrt2}\big(\sigma_z+\sigma_x\big)$, dejando que Bob mida con $B=\sigma_z$ y $b=\sigma_x$, vemos que

$$|\langle CHSH_1\rangle|=2\sqrt{2} > 2$$

También se puede demostrar que $2\sqrt{2}$ es el valor máximo posible alcanzable, incluso en el caso cuántico (línea de puntos y rayas en la gráfica).

$$$$

El hecho de que hayamos visto el incumplimiento de la desigualdad CHSH en un dispositivo real es significativo. Hace apenas una década, un experimento así habría tenido un gran impacto. Hoy en día, los dispositivos cuánticos se han vuelto significativamente mejores y estos resultados se pueden replicar fácilmente en hardware de última generación. 

Dado que podemos generar pares entrelazados con alta fidelidad y cada medición produce un resultado (es decir, ninguna partícula medida se "pierde"), hemos cerrado la brecha de detección en nuestros experimentos anteriores. Sin embargo, dada la distancia entre nuestros qubits (unos pocos $mm$) y el tiempo que se tarda en realizar una medición (del orden de $\mu$s), no podemos afirmar que hemos cerrado la brecha de [causalidad](https://en.wikipedia.org/wiki/Causality).
